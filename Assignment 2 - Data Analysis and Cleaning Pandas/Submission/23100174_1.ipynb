```python
correct_locations = fake_jobs['location'].str.contains('^\w+, \w+, \w+$', regex = True)  
fake_jobs = fake_jobs[correct_locations]                                #extracting rows which follow the correct location format(city, state, country)

fake_jobs['city'] = fake_jobs['location'].str.split(r',').str[0]        #extracting city from "location" values and making a separate column of city in dataframe
fake_jobs['state'] = fake_jobs['location'].str.split(r',').str[1]       #extracting state from "location" values and making a separate column of state in dataframe
fake_jobs['country'] = fake_jobs['location'].str.split(r',').str[2]     #extracting country from "location" values and making separate column of it in dataframe

fake_jobs = fake_jobs.drop('location', axis=1)                          #dropping the "location" column
```
